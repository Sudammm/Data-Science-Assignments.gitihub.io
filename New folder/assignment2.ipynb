```python
df['salary']=df['salary'].map({'low':0,'medium':0.5,'high':1})
df
```

```python
depart=pd.get_dummies(df['Department'],sparse=True,drop_first=False,prefix='depar')
```

```python
new_df=pd.concat([df,depart],axis=1)
new_df
```

```python
new_df.drop(columns=['Department','salary'],inplace=True)
new_df
```

```python
X=new_df.drop(columns=['left'])
X.columns
```

```python
y=new_df['left']
y
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
X_train
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```
